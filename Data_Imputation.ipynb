```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.impute import KNNImputer
```

```python
df_updated = pd.read_csv('df_combined.csv')
```

```python
df_updated_sample = df_updated[18252:19287]
```

```python
df_updated_sample
```

```python
df_updated_sample = df_updated_sample.drop('Unnamed: 0', axis=1)
```

```python
original_columns = df_updated_sample.columns.tolist()
```

```python
# iterative imputation with randomforest regressor (too computationally intensive, working out best sample sizes for imputation)

# separate the DataFrame into numerical and categorical parts
df_numerical = df_updated_sample.select_dtypes(include=['int64', 'float64'])
df_categorical = df_updated_sample.select_dtypes(exclude=['int64', 'float64'])

# qdjusted RandomForestRegressor within IterativeImputer
regressor = RandomForestRegressor(
    n_estimators=120,  # slightly more trees
    max_depth=None,    # no limit on the depth
    min_samples_leaf=2,  # more generalization
    max_features='sqrt',  # default
    random_state=0,    # dor reproducibility
    n_jobs=-1          # use all available CPUs
)

# Initialize the IterativeImputer with RandomForestRegressor
imputer = IterativeImputer(estimator=regressor, max_iter=10, random_state=0)

# Fit the imputer and transform the numerical DataFrame
df_numerical_imputed = imputer.fit_transform(df_numerical)
```

```python
df_numerical_imputed
```

```python
df_numerical_imputed = pd.DataFrame(df_numerical_imputed, columns=df_numerical.columns)
```

```python
df_categorical
```

```python
df_numerical_imputed.reset_index(drop=True, inplace=True)
df_categorical.reset_index(drop=True, inplace=True)
```

```python
df_combined = pd.concat([df_numerical_imputed, df_categorical], axis=1)
```

```python
df_combined
```

```python
df_imputed = df_combined.tail(11)
```

```python
df_imputed = df_imputed[original_columns]
```

```python
df_imputed.to_csv('df_imputed.csv')
```
